```python
# Softmax 함수 정의 (안정성을 위해 scipy 사용 추천, 없으면 numpy로 구현 가능)
from scipy.special import softmax
def apply_softmax(x):
    return softmax(x)

# groupby 후 transform을 사용하면 원본 인덱스를 유지하며 값을 변환해 채워넣습니다.
# 즉, 이미지 1장 내에 있는 모든 헤드들의 점수를 모아서 Softmax를 취합니다.
df_input['softmax_score'] = df_input.groupby(['object', 'action', 'filename'])['score'].transform(apply_softmax)

df_result = df_input.groupby(['layer', 'head'])['softmax_score'].mean().sort_values(ascending=False).reset_index()
```

```python
df_result.groupby(["layer"]).agg({"softmax_score": "mean"}).sort_values("softmax_score", ascending=False).reset_index().head(10)
```

```python
df_result.groupby(["layer", "head"]).agg({"softmax_score": "mean"}).sort_values("softmax_score", ascending=False).reset_index().head(10)
```
