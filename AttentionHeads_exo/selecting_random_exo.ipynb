```python
import pandas as pd
import os
import glob
import random

import pandas as pd
import os
import sys
current_dir = os.path.dirname(os.getcwd())
sys.path.append(current_dir)


import numpy as np

from file_managing import make_input_image
from config import AGD20K_PATH, model_name
AGD20K_PATH
# 1. 데이터프레임 로드 및 전처리
df = pd.read_pickle("target_df_w_description.pkl")
if 'description' in df.columns:
    df.drop('description', axis=1, inplace=True)

def fetch_random_exo_image(row):
    """
    행(row)의 action과 object 정보를 바탕으로 
    해당 디렉토리에서 랜덤한 이미지 경로를 반환합니다.
    """
    action = row['action']
    obj = row['object']
    
    # 디렉토리 경로 구성
    # 구조: {AGD20K_PATH}/Seen/trainset/exocentric/{action}/{object}
    search_dir = os.path.join(AGD20K_PATH, 'Seen', 'trainset', 'exocentric', action, obj)
    
    # 디렉토리가 존재하는지 확인
    if os.path.exists(search_dir):
        # 해당 폴더 내의 모든 jpg 파일 리스트 가져오기
        files = glob.glob(os.path.join(search_dir, "*.jpg"))
        
        if files:
            # 파일 목록 중 하나를 랜덤으로 선택
            selected_path = random.choice(files)
            
            # [옵션 1] 전체 경로를 저장하고 싶다면:
            return selected_path
            
            # [옵션 2] 파일명만 저장하고 싶다면 아래 주석을 해제하고 위를 주석 처리하세요:
            # return os.path.basename(selected_path)
            
    return None  # 경로가 없거나 파일이 없는 경우

# 3. 함수 적용 (각 행마다 실행)
# tqdm을 사용하면 진행 상황을 바(bar)로 볼 수 있어 편리합니다 (선택사항)
try:
    from tqdm import tqdm
    tqdm.pandas()
    df['random_exo_filename'] = df.progress_apply(fetch_random_exo_image, axis=1)
except ImportError:
    df['random_exo_filename'] = df.apply(fetch_random_exo_image, axis=1)

# 5. 피클로 저장
df.to_pickle("target_df_w_random_exo.pkl")
df
```
