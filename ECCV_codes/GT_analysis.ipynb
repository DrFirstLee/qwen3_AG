```python
import os

# 1. ê²½ë¡œ ì„¤ì • (ì‚¬ìš©ìë‹˜ì˜ ì„œë²„ í™˜ê²½ì— ë§ì¶¤)
base_path = "/home/DATA/AGD20K/Seen/testset/egocentric"

# 2. ì‚¬ìš©ìë‹˜ì´ ì„ ì •í•œ 32ê°œ Hard Pairs ì •ì˜
# (action, object) íŠœí”Œ ë¦¬ìŠ¤íŠ¸
hard_pairs = [
    # Category 1: Co-occurring
    ('brush_with', 'toothbrush'), ('hold', 'toothbrush'),
    ('cut_with', 'knife'), ('hold', 'knife'),
    ('cut_with', 'scissors'), ('hold', 'scissors'),
    ('hit', 'axe'), ('hold', 'axe'),
    ('hit', 'hammer'), ('hold', 'hammer'),
    ('drag', 'suitcase'), ('hold', 'suitcase'),
    ('lift', 'fork'), ('hold', 'fork'),
    ('drink_with', 'cup'), ('hold', 'cup'),
    ('drink_with', 'bottle'), ('hold', 'bottle'),
    
    # Category 2: Semantically Similar
    ('cut_with', 'knife'), ('stick', 'knife'),
    ('ride', 'bicycle'), ('sit_on', 'bicycle'), ('push', 'bicycle'),
    ('ride', 'motorcycle'), ('sit_on', 'motorcycle'),
    ('carry', 'skis'), ('jump', 'skis'),
    ('carry', 'snowboard'), ('jump', 'snowboard'),
    ('talk_on', 'cell_phone'), ('text_on', 'cell_phone'),
    ('kick', 'soccer_ball'), ('throw', 'soccer_ball'), ('catch', 'soccer_ball'),
    ('kick', 'rugby_ball'), ('throw', 'rugby_ball'), ('catch', 'rugby_ball')
]

# 3. ì¹´ìš´íŒ… ë¡œì§
total_images = 0
hard_pair_images = 0
pair_stats = {}

# ì§€ì›í•˜ëŠ” ì´ë¯¸ì§€ í™•ì¥ì
valid_extensions = ('.jpg', '.jpeg', '.png')

for action in os.listdir(base_path):
    action_dir = os.path.join(base_path, action)
    if not os.path.isdir(action_dir): continue
    
    for obj in os.listdir(action_dir):
        obj_dir = os.path.join(action_dir, obj)
        if not os.path.isdir(obj_dir): continue
        
        # í•´ë‹¹ ë””ë ‰í† ë¦¬ì˜ ì´ë¯¸ì§€ íŒŒì¼ ëª©ë¡ ì¶”ì¶œ
        images = [f for f in os.listdir(obj_dir) if f.lower().endswith(valid_extensions)]
        num_images = len(images)
        
        # ì „ì²´ ì´ë¯¸ì§€ ëˆ„ì 
        total_images += num_images
        
        # Hard Pairì— í•´ë‹¹í•˜ë©´ ëˆ„ì 
        if (action, obj) in hard_pairs:
            hard_pair_images += num_images
            pair_stats[f"{action}-{obj}"] = num_images

# 4. ê²°ê³¼ ì¶œë ¥
print("="*50)
print(f"ğŸ“Š AGD20K Seen Testset í†µê³„")
print("-"*50)
print(f"âœ… ì „ì²´ ì´ë¯¸ì§€ ìˆ˜: {total_images}")
print(f"ğŸ”¥ Hard Pair ì´ë¯¸ì§€ ìˆ˜ (ëª¨ìˆ˜): {hard_pair_images}")
print(f"ğŸ“ˆ ì „ì²´ ëŒ€ë¹„ Hard Pair ë¹„ì¤‘: {(hard_pair_images/total_images)*100:.2f}%")
print("-"*50)
print(f"ğŸš¨ ì´ {hard_pair_images}ì¥ ì¤‘ 255ì¥ì´ ì‹¤íŒ¨í–ˆë‹¤ë©´?")
print(f"ğŸ‘‰ ì‹¤ì§ˆì  ì‹¤íŒ¨ìœ¨: {(255/hard_pair_images)*100:.2f}%")
print("="*50)

# ìƒì„¸ ì •ë³´ ë³´ê³  ì‹¶ì„ ë•Œ ì£¼ì„ í•´ì œ
# for pair, count in pair_stats.items():
#     print(f"{pair}: {count}ì¥")b
```

```python
import os

# 1. ê²½ë¡œ ì„¤ì • (Unseen ê¸°ì¤€)
ego_root = "/home/DATA/AGD20K/Unseen/testset/egocentric"
gt_root = "/home/DATA/AGD20K/Unseen/testset/GT"

def count_files(root_dir, extensions):
    count = 0
    if not os.path.exists(root_dir):
        return 0
    
    for action in os.listdir(root_dir):
        action_path = os.path.join(root_dir, action)
        if not os.path.isdir(action_path): continue
        
        for obj in os.listdir(action_path):
            obj_path = os.path.join(action_path, obj)
            if not os.path.isdir(obj_path): continue
            
            # í•´ë‹¹ ë””ë ‰í† ë¦¬ ë‚´ ì§€ì›í•˜ëŠ” í™•ì¥ì íŒŒì¼ ì¹´ìš´íŠ¸
            files = [f for f in os.listdir(obj_path) if f.lower().endswith(extensions)]
            count += len(files)
    return count

# 2. ì‹¤í–‰
# ì›ë³¸ ì´ë¯¸ì§€ëŠ” ë³´í†µ .jpg, GTëŠ” .pngì¸ ê²½ìš°ê°€ ë§ìœ¼ë¯€ë¡œ ê°ê° ì„¤ì •
ego_count = count_files(ego_root, ('.jpg', '.jpeg', '.png'))
gt_count = count_files(gt_root, ('.png', '.jpg', '.jpeg'))

# 3. ê²°ê³¼ ì¶œë ¥
print("="*50)
print(f"ğŸ“‚ AGD20K [Unseen] Testset í†µê³„")
print("-"*50)
print(f"ğŸ–¼ï¸ ì›ë³¸ ì´ë¯¸ì§€(Egocentric) ì´ ê°œìˆ˜: {ego_count:5d} ì¥")
print(f"ğŸ­ ì •ë‹µ ë§ˆìŠ¤í¬(GT) ì´ ê°œìˆ˜:        {gt_count:5d} ì¥")
print("-"*50)

if ego_count == gt_count:
    print("âœ… í™•ì¸ ê²°ê³¼: ì´ë¯¸ì§€ì™€ GTì˜ ê°œìˆ˜ê°€ ì™„ë²½íˆ ì¼ì¹˜í•©ë‹ˆë‹¤.")
else:
    diff = abs(ego_count - gt_count)
    print(f"âš ï¸ ì£¼ì˜: ê°œìˆ˜ê°€ ì¼ì¹˜í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤! (ì°¨ì´: {diff}ì¥)")
    print("   í‰ê°€ ì‹œ GTê°€ ì—†ëŠ” ì´ë¯¸ì§€ëŠ” ì œì™¸ë  ìˆ˜ ìˆìœ¼ë‹ˆ í™•ì¸ì´ í•„ìš”í•©ë‹ˆë‹¤.")
print("="*50)
```
