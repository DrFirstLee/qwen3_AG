```python
import numpy as np

# 1. Category A Î¶¨Ïä§Ìä∏ Ï†ïÏùò
category_a_list = [
    "brush_with_toothbrush", "hold_toothbrush",
    "cut_with_knife", "hold_knife",
    "cut_with_scissors", "hold_scissors",
    "hit_axe", "hold_axe",
    "hit_hammer", "hold_hammer",
    "lift_fork", "hold_fork",
    "drink_with_cup", "sip_cup", "pour_cup", "hold_cup",
    "drink_with_bottle", "pour_bottle", "hold_bottle",
    "drink_with_wine_glass", "sip_wine_glass", "pour_wine_glass", "hold_wine_glass",
    "hit_baseball_bat", "hold_baseball_bat",
    "hit_tennis_racket", "swing_tennis_racket", "hold_tennis_racket"
]

# 2. Ï°∞Í±¥ ÎßàÏä§ÌÅ¨ ÏÉùÏÑ± (case_idÍ∞Ä Î¶¨Ïä§Ìä∏Ïóê ÏûàÏúºÎ©¥ True)
mask = df_pkl['case_id'].isin(category_a_list)

# 3. Í∞í ÏÑ†ÌÉù (np.where)
# Ï°∞Í±¥Ïù¥ True(Category A)Î©¥ 'no_coocur' Í∞í ÏÇ¨Ïö©, ÏïÑÎãàÎ©¥ 'random' Í∞í ÏÇ¨Ïö©
df_pkl['mixed_metric'] = np.where(mask, df_pkl['no_coocur'], df_pkl['random'])

# 4. ÌèâÍ∑† Í≥ÑÏÇ∞
final_average = df_pkl['mixed_metric'].mean()

print(f"‚úÖ Final Mixed Average: {final_average:.4f}")

# (ÏÑ†ÌÉùÏÇ¨Ìï≠) Î™á Í∞úÏî© ÏÇ¨Ïö©ÎêòÏóàÎäîÏßÄ ÌôïÏù∏
count_a = mask.sum()
count_others = (~mask).sum()
print(f"üëâ Used 'no_coocur' for {count_a} cases (Category A)")
print(f"üëâ Used 'random' for {count_others} cases (Others)")
```
